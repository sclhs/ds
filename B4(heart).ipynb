```python
#countplot
sns.countplot(x="Sex", data=df, palette="mako_r")
plt.title("SEX COUNT", size=12, c='b')
plt.show()
```

```python
#DISTPLOT
plt.title("DISTRIBUTION PLOT")
sns.distplot(df["Cholesterol"])
plt.show()
```

```python
#SCATTER PLOT
plt.figure(figsize=(16,10))
sns.scatterplot([df.RestingBP, df.MaxHR])
plt.show()
```

```python
#GROUPED BARPLOT
plt.figure(figsize=(15,8))
width = 0.4
plt.bar(df.Age[:200]-0.2, df.Cholesterol[:200], width, color="orange")
plt.bar(df.Age[:200]+0.2, df.MaxHR[:200], width, color="green")
plt.legend(['Cholesterol', 'MaxHR'])
plt.show()
```

```python
#pandas.crosstab()
# Crosstabs are used for categorical data
# Crosstabs are used to analyze the relationship between two categorical variables, 
#while pivot tables can analyze the relationships between multiple variables, both categorical and numerical.
```

```python
pd.crosstab(df.Age, df.HeartDisease).plot(kind='bar', figsize=(16,8))
plt.xticks(rotation='horizontal', c='r')
plt.yticks(c='r')
plt.title("CROSS TAB - AGE vs HEARTDISEASE")
plt.xlabel("AGE", size=18)
plt.show()
```

```python
#HISTOGRAM
sns.histplot(df['Age'], kde=True)
```

```python
hist_list=['Age','RestingBP','Cholesterol','FastingBS', 'MaxHR', 'Oldpeak']
hist_list
```

```python
df.hist(hist_list, figsize=(20,20), grid=False, bins=20, color='violet', legend=True)
plt.show()
```

```python
#BOXPLOT
df.boxplot(hist_list, figsize=(11,10), color='maroon', grid=False)
plt.show()
```
